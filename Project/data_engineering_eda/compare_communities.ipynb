```python
#import dependencies
import pandas as pd
import json
import csv
```

```python
#read the csv
file = "resources/chi_movies_in_the_park_2014_2019.csv"
raw_df = pd.read_csv(file)

raw_df.head()
```

```python
df=raw_df[["title","park","date","community"]]
df.head()
df.info()
```

```python
with open('resources/chicago-community-areas.geojson', "r") as read_file:
    data = json.load(read_file)

data
```

```python
communities = []
for feature in data['features']:
    comm = feature['properties']['community']
    communities.append(comm)
print(communities)
```

```python
test_list = list(df.community)
test_list = [str(x).strip() for x in test_list]

communities = [x.strip() for x in communities]
print(test_list[1])

matches = []
no_match = []

match_count = 0
no_match_count = 0
index = -1

for i in test_list:
    index +=1
    if i in communities:
        matches.append(test_list[index])
        match_count += 1
    else:
        no_match.append(test_list[index])
        no_match_count += 1

print(match_count)
print(no_match_count)
```
